4. Примеры индексирования и срезов

```python
#Получение одного элемента массива
a = np.arange(15)
print(a)
a[1]
```

```python
#Срез с 2 по 5 элемент
a[2:6]
```

```python
#Выбрать конкретные элементы по индексу
a[[2, 6]]
```

```python
#Выбрать конкретные элементы по индексу, можно любой порядок
a[[7, 1, 4]]
```

```python
#Можно повторять значения
a[[1, 2, 1, 7, 1]]
```

```python
#Взять последний элемент
a[-1]
```

Работа с многомерными массивами

```python
b = np.array([[1, 1.5, 2, 2.5],
              [3, 3.5, 4, 4.5],
              [5, 5.5, 6, 6.5]])
b
```

```python
#Обращение к элементу
print(b[0, 1])
print(b[(0, 1)])
#Срез подмассива
print(b[[0, 1]])
```

```python
#Срез подмассива
b[[0, 2], 0]
#Взятие конкретных элементов
b[[0, 2], [1, 3]]
```

```python
#Взятие двумерного подмассива
b[1:3, 0:2]
```

```python
#Взятие одномерного подмассива
b[1:2, 0:2]
```

```python
#Взять конкретный столбец целиком
b[:, 1]
```

```python
#Взять строку целиком
print(b[1])
print(b[1, :])
```

```python
#Взять весь массив
b[:]
```

```python
#Выбрать первые 2 строки, указывается идекс, который уже не входит
b[:2]
```

```python
#Выбрать первые 2 строки, указывается идекс, который уже не входит
# :: - все
b[:2, ::]
```

```python
#Выбрать первые 2 строки и столбцы со 2 индекса
b[:2, 2:]
```

```python
#Берет первую строку из получившегося среза
b[0:2][0]
```

```python
#Создание массива
c = np.empty((2, 3, 3, 3))
#Выбрать элементы из полседней оси
c[..., 2]
```

5. Примеры векторных операций и базовых статистик

```python
a = np.arange(2, 20, 2)
a
```

```python
b = np.arange(10, 100, 10)
b
```

```python
#Поэлментное сложение массивов, важно, чтобы были одной длины
c = a + b
c
```

```python
#Поэлементное умножение массивов
c = a*b
c
```

```python
#Вычисление средней величины
m = np.mean(c)
m
```

```python
#Вычисление медианы
m = np.median(c)
m
```

```python
#Минимальное и максимальное значения
print(np.min(a))
print(np.max(a))
```

```python
#Вычисление дисперсии
d = np.var(a)
d
```

```python
#Найти индекс числа с минимальным значением
m_i = np.argmin(a)
m_i
```
